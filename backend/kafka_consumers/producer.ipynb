```python
from confluent_kafka import Producer
import time

# Kafka Producer configuration
conf = {
    'bootstrap.servers': 'localhost:9092'
}

# Create a Producer instance
producer = Producer(conf)

# Delivery report handler
def delivery_report(err, msg):
    """ Called once for each message produced to indicate delivery result.
        Triggered by poll() or flush(). """
    if err is not None:
        print(f"Message delivery failed: {err}")
    else:
        print(f"Message delivered to {msg.topic()} [{msg.partition()}]")

# Send a few test messages to the Kafka topic 'test-topic'
for i in range(5):
    message = f'Hello Kafka! Message {i}'
    producer.produce('test-topic', key=str(i), value=message, callback=delivery_report)
    # Wait up to 1 second for events to be triggered
    producer.poll(1)
    time.sleep(1)  # Sleep for a bit between messages to simulate work

# Flush any remaining messages in the queue
producer.flush()
```
